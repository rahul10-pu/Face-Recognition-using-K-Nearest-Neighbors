**Let us consider a binary image named checker bilevel (PNG format) from scikit-image library available within Anaconda. An image having values either 0 for black and 255/1 for white is termed as a binary image.**

```
#import and save the image in a variable named img
import os.path
from skimage.io import imread
from skimage import data_dir
img = imread(os.path.join(data_dir, 'checker_bilevel.png'))
```

Next, we visualize our image using Matplotlib* imshow() function as shown with colormap equals Greys.

```
import matplotlib.pyplot as plt
plt.imshow(img, cmap = 'Greys')
```

```
''' Image stored as numpy object (2D Matrix) '''
print(type(img))        # <class 'numpy.ndarray'>
''' Number of image dimensions '''
print(img.ndim)         # 2
''' Shape of the image (rows, columns) '''
print(img.shape)        # (10, 10)
''' Number of total elements in the image '''
print(img.size)         # 100 
''' Size of per element (in bytes) '''
print(img.itemsize)     # 1
''' Size of complete image (in bytes) '''
print(img.nbytes)       # 100
```

```
# Transpose
img_t = img.T
plt.figure()
plt.imshow(img_t, cmap = 'Greys')
# Reshape
img_reshape = img.reshape(5, 20)
plt.figure()
plt.imshow(img_reshape, cmap = 'Greys')
# Sort
img_srt = img.copy()
img_srt.sort(axis = 0) 
plt.figure()
plt.imshow(img_srt, cmap = 'Greys')
# Compression
img_cmp = img.copy()
img_cmp = img_cmp.compress([True,False,True,0,1,1,1,0,0,1],axis = 0)
plt.figure()
plt.imshow(img_cmp, cmap = 'Greys')
print(img_cmp.shape)
```
